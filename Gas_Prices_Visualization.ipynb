```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
gas=pd.read_csv('/content/gas_prices.csv')
```

```python
gas
```

```python
gas_USA=gas['USA']
gas_Canada=gas['Canada']
gas_sK=gas['South Korea']
#plt.figure(figsize=(8,5))

plt.plot(gas.Year,gas_USA, 'b.-',label='USA' )
plt.plot(gas.Year,gas_Canada,'g.-',label='Canada')
plt.plot(gas.Year,gas_sK,'r.-',label='South Korea')
plt.xlabel("Year")
plt.ylabel("US Dollar")
plt.title(" Gas  Prices")
plt.xticks(gas['Year'][::3].tolist()+[2011])
plt.yticks([2,4,6,8,10])
plt.legend()
```

```python
for country in gas:
  if country != 'Year':
    plt.plot(gas['Year'], gas[country], marker='.',label=country)
    plt.xlabel("Year")
plt.ylabel("US Dollar")
plt.title(" Gas  Prices", fontdict={'fontweight':'bold', 'fontsize':22})
plt.xticks(gas['Year'][::3].tolist()+[2011])
plt.yticks([2,4,6,8,10])

plt.legend()

plt.savefig("Gas_Price_Figure.png",dpi=200)
```
